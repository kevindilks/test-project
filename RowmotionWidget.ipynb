```python
x_widget=widgets.IntSlider(min=0,max=len(P.rowmotion_orbits())-1,step=1,value=0)
y_widget=widgets.IntSlider(min=0,max=0,step=1,value=0)
```

```python
def update_y(*args):
    y_widget.max = len(P.rowmotion_orbits()[x_widget.value])-1
x_widget.observe(update_y, 'value')
```

```python
def printer(P,x,y):
    return(P.plot(vertex_colors={'blue':P.rowmotion_orbits()[x][y]}))
```

```python
interact(printer,P=fixed(P),x=x_widget,y=y_widget)
```
